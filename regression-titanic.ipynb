### Create a subset with the passangers that survived

```python
# select the passangers that survived
titanic_survived = titanic.loc[titanic['Survived'] == 1]
titanic_survived
```

### Select corresponding features

```python
# X = titanic_survived['Pclass'].values
# y = titanic_survived['Survived'].values

X = titanic['Age'].values
y = titanic['Pclass'].values

# X = titanic['Pclass'].values
# y = titanic['Survived'].values

# Transform from 1D to 2D, with all rows as needed(-1)
X = X.reshape(-1,1)
# X
```

### Create trains

```python
# X_feature - Pclass
# y_output - Survived
x_train, x_test, y_train, y_test = train_test_split(X,y,train_size=0.8,test_size=0.2,random_state=100)
print(f"X_train shape {x_train.shape}")
print(f"y_train shape {y_train.shape}")
print(f"X_test shap {x_test.shape}")
print(f"y_test shape {y_test.shape}")
print(y_test)
```

### Create plot to show

```python
#  had to add the nexy line to maje it work
import matplotlib.pyplot as plt

%matplotlib inline
plt.scatter(x_train,y_train,color='purple')
plt.xlabel('Cabin Class')
plt.ylabel('Survived')
plt.title('Does the cabin class affected who survived?')
plt.show()
```
